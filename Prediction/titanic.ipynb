# Setting up the Enviroment

```python
import numpy as np
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import LabelEncoder
```

```python
train = pd.read_csv('train.csv')
test = pd.read_csv('test.csv')
```

```python
train.head()
```

```python
train.shape
```

```python
train.describe()
```

# Handling Missing Values

```python
train.isnull().sum()
```

```python
train.drop(columns=['Cabin'],inplace=True)
```

```python
test.drop(columns=['Cabin'],inplace=True)
```

```python
train.isnull().sum()
```

```python
train['Age'].fillna(train['Age'].mean(),inplace=True)
```

```python
train.isnull().sum()
```

# Handling the Data Values for RFC

```python
print(train.dtypes)
# Display columns with object (string) data types
print(train.select_dtypes(include=['object']).columns)
```

```python
train.drop(columns=['Name'],inplace=True)
train.drop(columns=['Ticket'],inplace=True)
test.drop(columns=['Name'],inplace=True)
test.drop(columns=['Ticket'],inplace=True)
```

```python
print(train.dtypes)
print(test.dtypes)
```

```python
train['Sex'] = train['Sex'].map({'male':0,'female':1})
test['Sex'] = test['Sex'].map({'male':0,'female':1})
```

```python

train['Embarked']=train['Embarked'].map({'S':0,'C':1,'Q':2})

test['Embarked']=test['Embarked'].map({'S':0,'C':1,'Q':2})
```

```python
print(train.dtypes)
print(test.dtypes)
```

```python
X = train.drop(columns=['Survived'],axis=1)
y = train['Survived']
```

```python
test.isnull().sum()

```

```python
test['Age'].fillna(test['Age'].mean(),inplace=True)
test['Fare'].fillna(test['Fare'].mean(),inplace=True)
```

```python
test.isnull().sum()

```

# Applying RandomForestClassifier

```python
model = RandomForestClassifier(n_estimators=100,random_state=42)
model.fit(X,y)
```

```python
X_test = test.drop(columns=['Survived'], axis=1, errors='ignore')

predictions = model.predict(X_test)
```

# Score/Rating :

```python

train_predictions = model.predict(X)


accuracy = accuracy_score(y, train_predictions)
print('Accuracy on the training set:', accuracy)
```

```python
submission = pd.DataFrame({
    'PassengerId': test['PassengerId'],
    'Survived': predictions
})

submission.to_csv('submission.csv', index=False)
```
